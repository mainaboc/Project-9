```python
df2_2 = df_mental_2019.sort_values("Prevalence - Mental disorders - Sex: Both - Age: Age-standardized (Percent)", ascending = False)
```

```python
#Merging the 10 highest and lowest HDIs with heir respective proportion of mental health disorders.
df_merged2 = df_merge1_1.merge(mental, on="Entity")
```

```python
#Creating a new dataframe in order to plot without extreme outliers. 
df_scatter = df_merged[df_merged.Entity != "World"]
```

```python
#Ploting a scatterplot to determine a correlation.
plt.scatter(df_scatter["Human Development Index"], df_scatter["Prevalence - Mental disorders - Sex: Both - Age: Age-standardized (Percent)"])
plt.xlabel("Human Development Index")
plt.ylabel("% of population with mental health disorders")
plt.title("Worldwide")
```

```python
#Scatterplot for the 10 lowest and highest HDIs in order to better determine a correlation.
plt.scatter(df_merged2["Human Development Index"], df_merged2["Prevalence - Mental disorders - Sex: Both - Age: Age-standardized (Percent)"])
plt.xlabel("Human Development index")
plt.ylabel("% population with mental health disorders")
plt.title("10 lowest and 10 highest HDI countries")
```
